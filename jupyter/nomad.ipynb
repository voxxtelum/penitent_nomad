```python
import pandas as pd

container_locations = pd.read_csv('../store_spider/output/container_locations_mid.csv')

def split_loc(location):
  loc_split = location.split(',', 1)
  # print(loc_split)
  city = loc_split[0].strip()
  state = loc_split[1].split('\xa0')[0].strip()
  zip = loc_split[1].split('\xa0')[1].strip()
  return [city, state, zip]
  
clean = []

for row in container_locations.iterrows():
  if row[1][0] == row[1][2]:
    split = split_loc(row[1][1])
    row_add = {'store_name': split[0], 'store_add1':row[1][0], 'store_add2': '', 'store_city': split[0], 'store_state': split[1], 'store_zip': split[2]}

    clean.append(row_add)
  else:
    split = split_loc(row[1][2])
    row_add = {'store_name': split[0], 'store_add1':row[1][0], 'store_add2': row[1][1], 'store_city': split[0], 'store_state': split[1], 'store_zip': split[2]}

    clean.append(row_add)

  
  
# print(clean)
# container_locations.head()

clean_frame = pd.DataFrame(clean)
display(clean_frame)
```

```python
clean_frame.to_csv('../output/container_locations_clean.csv')
```
